```python
import pandas as pd
import numpy as np
```

```python
list_data = [ np.random.random() for _ in range(200) ]
```

```python
series = pd.Series(list_data)
```

```python
print(series)
```

```python
#iterating through series
#Element-wise
for element in series:
    print(element)
```

```python
#Index-wise
for i in range(len(series)):
    print(series[i])
```

```python
series[0]
```

```python
#minimum value
min_value = series.min()
print(min_value)

#index of minimum value
min_index = series.argmin()
print(min_index)

#maximum
max_value = series.max()
print(max_value)
max_index = series.argmax()
print(max_index)
```

```python
mean = series.mean()
print(mean)
```

```python
median = series.median()
print(median)
```

```python
#mode - the most frequent value
mode = series.mode()
print(mode)
```
